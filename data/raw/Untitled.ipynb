# Headturn

```python
df = pd.read_csv("Clip_data.csv")
```

```python
print("Shape:", df.shape)
print("\nColumns (first30):\n",df.columns[:30])
print("\nInfo:")
df.info()
print("\nDescribe numeric columns:")
print(df.describe().T.head(20))
print("\nMissing values per column:")
print("\nHead of dataset:")
print(df.head())
print("\nTail of dataset")
print(df.tail())
```

```python
# Search for columns related to hands, wrists and arms
hand_cols = [c for c in df.columns if any(k in c.lower() for k in["hand","wrist","arm"])]
print(hand_cols)
```

```python
len(hand_cols)
```

```python
# Looking for wheel socket or IK target
wheel_cols = [c for c in df.columns if "wheel" in c.lower()]
print("Wheel-related columns:", wheel_cols)
```

```python
# Quick stats for hand/wrist signals
print(df[hand_cols].describe())
```

```python
# Calculate std (spread) and range (max-min) for each hand-related column
movement_stats = df[hand_cols].agg(['std', lambda x: x.max() - x.min()]).T
movement_stats.columns = ['std', 'range']

# Sort by standard deviation (biggest movers at top)
movement_stats_sorted = movement_stats.sort_values('std', ascending=False)

# Show top 20 candidates
print(movement_stats_sorted)
```

```python
movement_stats_sorted.to_csv("hand_movement_stats.csv")

print("Exported to hand_movement_stats.csv")
```

```python


selected_cols = [
    "RightWrist_flexion", "RightWrist_adduction",
    "RightWrist_flexion_velocity", "RightWrist_adduction_velocity",
    "LeftWrist_flexion", "LeftWrist_adduction",
    "LeftWrist_flexion_velocity", "LeftWrist_adduction_velocity"
]

# Keep only available columns + timestamp
steering_df = df[["Timestamp"] + [c for c in selected_cols if c in df.columns]].copy()

# --- Create Steering Proxy ---
# Combine right wrist flexion + adduction (adjust weights if needed)
if "RightWrist_flexion" in steering_df.columns and "RightWrist_adduction" in steering_df.columns:
    steering_df["steer_proxy"] = (
        0.7 * steering_df["RightWrist_flexion"].fillna(0) +
        0.3 * steering_df["RightWrist_adduction"].fillna(0)
    )
elif "RightWrist_flexion" in steering_df.columns:
    steering_df["steer_proxy"] = steering_df["RightWrist_flexion"].fillna(0)
else:
    steering_df["steer_proxy"] = steering_df["RightWrist_adduction"].fillna(0)

# Center by subtracting baseline (first ~0.5s or 30 frames)
baseline = steering_df["steer_proxy"].iloc[:30].median()
steering_df["steer_proxy"] = steering_df["steer_proxy"] - baseline

# --- Label the Data ---
th = 10.0  # threshold in degrees (adjust if needed)
labels = np.full(len(steering_df), "neutral", dtype=object)
labels[steering_df["steer_proxy"] > th] = "turn_right"
labels[steering_df["steer_proxy"] < -th] = "turn_left"
steering_df["steer_label"] = labels

# --- Save to CSV ---
steering_df.to_csv("steering_signals.csv", index=False)
print("Exported steering_signals.csv with", len(steering_df.columns), "columns (including proxy + label)")
```
